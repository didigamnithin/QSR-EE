```python
import os
import pandas as pd

# Root directory where your folders (QSR Executive Enterprises-...) are located
root_dir = "."

# Data containers
df_united_list = []
df_ads_list = []
df_offers_list = []

# Loop through folders
for folder in os.listdir(root_dir):
    folder_path = os.path.join(root_dir, folder)
    
    if os.path.isdir(folder_path) and folder.startswith("QSR"):
        print(f"\n📂 Processing folder: {folder}")

        # Find files inside the folder
        for file in os.listdir(folder_path):
            file_path = os.path.join(folder_path, file)

            # united file (UUID style)
            if file.endswith(".csv") and "ads" not in file and "offers" not in file:
                df = pd.read_csv(file_path)
                df["folder_name"] = folder
                df_united_list.append(df)
                print(f"  🟢 United file: {file} → {df.shape[0]} rows × {df.shape[1]} cols")

            # ads file
            elif file == "ads-campaigns-list.csv":
                df = pd.read_csv(file_path)
                df["folder_name"] = folder
                df_ads_list.append(df)
                print(f"  🔵 Ads file: {file} → {df.shape[0]} rows × {df.shape[1]} cols")

            # offers file
            elif file == "offers-campaigns.csv":
                df = pd.read_csv(file_path)
                df["folder_name"] = folder
                df_offers_list.append(df)
                print(f"  🟣 Offers file: {file} → {df.shape[0]} rows × {df.shape[1]} cols")

# Combine all data
df_united = pd.concat(df_united_list, ignore_index=True) if df_united_list else pd.DataFrame()
df_ads = pd.concat(df_ads_list, ignore_index=True) if df_ads_list else pd.DataFrame()
df_offers = pd.concat(df_offers_list, ignore_index=True) if df_offers_list else pd.DataFrame()

# Save to CSV
df_united.to_csv("all_united.csv", index=False)
df_ads.to_csv("all_ads.csv", index=False)
df_offers.to_csv("all_offers.csv", index=False)

# Print totals
print("\n✅ Combined DataFrames created:")
print(f"   🟢 all_united.csv → {df_united.shape[0]} rows × {df_united.shape[1]} cols")
print(f"   🔵 all_ads.csv → {df_ads.shape[0]} rows × {df_ads.shape[1]} cols")
print(f"   🟣 all_offers.csv → {df_offers.shape[0]} rows × {df_offers.shape[1]} cols")
```
